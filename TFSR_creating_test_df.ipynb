```python
test_files = os.listdir('test/audio') #storing the names of all the test_files in this audio folder.
```

```python
test_files[0:4]
```

```python
data = []
sampling_rate = []
names = []
features = []

try:
  for i  in test_files:
    data1, sampling_rate1 = librosa.load('test/audio/' + i, res_type = 'kaiser_fast')

    mfccs = np.mean(librosa.feature.mfcc(y = data1, sr = sampling_rate1, n_mfcc = 40).T, axis = 0)

    data.append(data1)
    sampling_rate.append(sampling_rate1)
    features.append(mfccs)
    names.append(i)
    
except:
  print("Error in ", i)
```

```python
test_df = pd.DataFrame([names, data, sampling_rate, features]).T
test_df.columns = ['names', 'data', 'sampling_rate', 'features']
test_df.head()
```

```python
test_df.to_csv("test_df.csv", index = None)
```
